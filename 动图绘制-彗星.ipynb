```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

plt.rcParams['animation.embed_limit'] = 2**64

t = np.linspace(0, 2*np.pi,100)
x = 2*np.sin(2*t)
y = 5*np.cos(3*t)

fig = plt.figure(figsize=(8,6.4)) # 建立图像画布
ax = fig.add_subplot(111, autoscale_on=True, xlim=(-np.max(x)-np.max(x)/8,np.max(x)+np.max(x)/8), ylim=(-np.max(y)-np.max(y)/8,np.max(y)+np.max(y)/8))
ax.set_xlabel('position')

# 设置小球
ball, = ax.plot([],[],linestyle='None',marker='o', markersize=10,markeredgecolor='k', color='orange',markeredgewidth=2)
line, = ax.plot([],[],linestyle='--',color='blue')

def init():
    ball.set_data(x[0],y[0])
    line.set_data(x[0],y[0])
    return ball,line

def animate(i):
    ball.set_data([x[i]],[y[i]])
    line.set_data([x[:i]],[y[:i]])
    return ball,line

animator = animation.FuncAnimation(fig, animate, np.arange(1, len(t)), interval=30, blit=False, init_func=init)


HTML(animator.to_jshtml())
```
