```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

plt.rcParams['animation.embed_limit'] = 2**64

x = np.linspace(0, 6.5, 51, endpoint=True)
y = np.sin(x)

fig, ax = plt.subplots()
line, = ax.plot(x, y, color='red')

def update(num, x, y, line):
    line.set_data(x[:num], y[:num])
    line.axes.axis([0, 7, -1.1, 1.1])
    return line,

animator = animation.FuncAnimation(fig, update, len(x), fargs=[x, y, line], interval=25, blit=False)

HTML(animator.to_jshtml())
```
